```python
import regex as re

def get_locus(x):
  x=str(x)
  y = re.search(r"OGR47_\d\d\d\d\d",x)
  try:
      out = x[y.span()[0]:y.span()[1]]
  except:
      out = "None"
  return out

def sub_fig(string):
  string2= re.sub(r"fig\|","",string)
  return string2
```

```python
df3=df2.drop(["dn1","dn1","dn2","dn3","dn4","dn5","dn6","dn7","dn8","dn9","dn10"], axis = 1)
df3=df3.drop_duplicates(subset=['parvus_id'],keep='first').reset_index(drop=True)
df3=df3.dropna()
df3
```

```python
df3["mjc1_id"]= df3["mjc1_id"].apply(lambda x: get_locus(x))
df3["parvus_id"]= df3["parvus_id"].apply(lambda x: sub_fig(x))
df3
```

```python
import regex as re
def sub_dash_dot(string):
  pattern = "\."
  string2= re.sub(pattern,"__SBML_DOT__",string)
  return string2
```

```python
df3["parvus_id2"]= df3["parvus_id"].apply(lambda x: sub_dash_dot(x))
df3
```

```python
df3.to_csv("parsing_parvus.txt", index=False)
df3=pd.read_csv("parsing_parvus.txt", header=0)
df3
```
